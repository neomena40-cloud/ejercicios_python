El Acceso a la Montaña Rusa
Para subir, hay reglas estrictas:
Seguridad: Medir más de 140 cm Y no tener problemas cardíacos.Vía Rápida: Si tienes el "Pase VIP", entras sin importar lo anterior (pero siempre que midas más de 120 cm por seguridad básica).

```python
estatura = 150
tiene_pase_vip = False
problemas_cardiacos = False
if estatura >=140 and not problemas_cardiacos:
  print("Puedes subir a la montaña rusa")
elif tiene_pase_vip and estatura >= 120:
  print("Puedes subir a la montaña rusa")
elif estatura <120 or problemas_cardiacos:
  print("No puedes subir a la montaña rusa")
else:
  print("No puedes subir a la montaña rusa")
```

```python
def acceso_montana_rusa():
    # Pedimos datos al usuario
    estatura = int(input("¿Cuál es tu estatura? (en cm) "))
    problemas = input("¿Tienes problemas cardíacos? (True/False): ")
    vip = input("¿Tienes pase VIP? (True/False): ")

    # Condiciones
    if vip == "True" and estatura >= 120:
        print("Acceso permitido por vía rápida 🎢")

    elif estatura > 140 and problemas == "False":
        print("Acceso permitido 🎉")

    else:
        print("Acceso denegado ❌")

# Llamamos a la función
acceso_montana_rusa()
```
