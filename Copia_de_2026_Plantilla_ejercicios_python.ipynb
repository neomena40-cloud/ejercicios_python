**Buscador de Números**

**Enunciado:** Dada la lista numeros = [10, 20, 30, 40, 50, 20, 60].

1. Encuentra en qué índice se encuentra el número 30.

2. Cuenta cuántas veces aparece el número 20.

3. Invierte el orden de la lista completa.

```python
# Lista de números
numeros = [10, 20, 30, 40, 50, 20, 60]

# Encontrar el índice del número 30
indice_30 = numeros.index(30)

# Contar cuántas veces aparece el número 20
cantidad_20 = numeros.count(20)

# Invertir el orden de la lista
numeros.reverse()

print("Índice de 30:", indice_30)
print("Cantidad de 20:", cantidad_20)
print("Lista invertida:", numeros)
```

**El Gestor de Invitados**

**Enunciado:** Tienes una lista inicial de invitados: invitados = ["Ana", "Juan", "Pedro"].

* Añade a "Maria" al final de la lista.

* Añade a "Luis" en la segunda posición (índice 1).

* Elimina a "Juan" de la lista usando su nombre.

* Imprime cuántos invitados hay en total usando una función.

```python
# Lista inicial de invitados
invitados = ["Ana", "Juan", "Pedro"]

# Añadir a "Maria" al final
invitados.append("Maria")

# Añadir a "Luis" en la segunda posición (índice 1)
invitados.insert(1, "Luis")

# Eliminar a "Juan" usando su nombre
invitados.remove("Juan")

# Imprimir cuántos invitados hay en total
total = len(invitados)

print("Lista final:", invitados)
print("Total de invitados:", total)
```

```python

nombres = ["Ana", "Sebastian", "Luz", "Alejandra"]
for nombre in nombres:
    if len(nombre) > 5:
        print(nombre)
```

**4. La cuenta atrás (While)**

Enunciado: Crea una variable energia = 5. Mientras la energía sea mayor a 0, imprime "Corriendo..." y resta 1 a la energía. Al final imprime "¡Me cansé!".

```python
# Variable inicial de energía
energia = 5

# Mientras la energía sea mayor que 0
while energia > 0:
    print("Corriendo...")
    energia -= 1  # Restar 1 a la energía en cada vuelta

# Cuando sale del bucle
print("¡Me cansé!")
```

**5. Filtrar pares**

Enunciado: Recorre los números del 1 al 10. Si el número es par, imprímelo. (Tip: un número es par si numero % 2 == 0).

```python
# Recorrer los números del 1 al 10
for numero in range(1, 11):
    # Verificar si el número es par
    if numero % 2 == 0:
        print(numero)
```

**Enunciado:**

Crea un diccionario con 3 colores en inglés y su traducción al español.

*   Usa un bucle for para imprimir solo las claves (los nombres en inglés).
*   Usa un bucle for para imprimir frases como: "The color [key] in Spanish is [value]".





```python
# Crear el diccionario con colores en inglés y su traducción al español
colores = {
    "red": "rojo",
    "blue": "azul",
    "green": "verde"
}

# 1. Imprimir solo las claves (los nombres en inglés)
for color in colores:
    print(color)

# 2. Imprimir frases con clave y valor
for color, value in colores.items():
    print(f"El color {color} en español es {value}")
```

**Enunciado:** Dada la palabra **"programacion"**, crea un diccionario que cuente cuántas veces aparece cada letra.

**Pista:** Usa un bucle para recorrer la palabra y el método .get() para inicializar el contador de cada letra.

```python
# Palabra dada
palabra = "programacion"

# Diccionario vacío para contar letras
contador = {}

# Recorrer cada letra de la palabra
for letra in palabra:
    # .get(letra, 0) devuelve el valor actual o 0 si no existe
    contador[letra] = contador.get(letra, 0) + 1

print(contador)
```

**El Carrito de la Compra**

**Enunciado:** Crea una lista vacía llamada carrito.

1. Pide al usuario que escriba 3 productos y añádelos uno a uno a la lista.

2. Muestra la lista ordenada alfabéticamente.

3.  Elimina el último producto que se añadió.

```python
# 1. Crear una lista vacía llamada carrito
carrito = []

# Pedir 3 productos al usuario y añadirlos a la lista
for i in range(3):
    producto = input("Escribe un producto: ")
    carrito.append(producto)

# 2. Mostrar la lista ordenada alfabéticamente
carrito.sort()
print("Carrito ordenado:", carrito)

# 3. Eliminar el último producto añadido
carrito.pop()
print("Carrito después de eliminar el último producto:", carrito)
```

```python
def gestionar_inventario():
    print("--- 🍔 BIENVENIDO A BURGER-PYTHON 🍔 ---")

    while True: # Bucle de recarga (el restaurante sigue abierto)
        inventario = 100
        print(f"\n[SISTEMA] Inventario recargado. Tienes {inventario} hamburguesas listas.")

        # Bucle de ventas (mientras haya stock)
        while inventario > 0:
            if inventario <= 10:
                print("🚨 ¡CUIDADO! Solo quedan", inventario, "hamburguesas.")
            try:
                pedido = int(input("\n¿Cuántas hamburguesas quiere el cliente? "))

                if pedido < 0:
                    print("❌ Error: No puedes pedir cantidades negativas.")
                elif pedido > inventario:
                    print(f"❌ No tenemos suficiente. Solo quedan {inventario}.")
                else:
                    inventario -= pedido
                    print(f"✅ Venta exitosa. Quedan {inventario} en stock.")
            except ValueError:
                print("❌ Por favor, ingresa un número válido.")
        # Cuando el inventario llega a 0, salimos del bucle interior y venimos aquí:
        print("\n--- 🛑 EL STOCK SE HA AGOTADO ---")
        respuesta = input("¿Deseas recargar el inventario para seguir vendiendo? (si/no): ").lower()

        if respuesta != "si":
            print("Cerrando el sistema. ¡Gracias por usar Burger-Python!")
            break # Rompe el bucle exterior y termina el programa
# Llamamos a la función para empezar
gestionar_inventario()
```

```python
import random

print("🎲 JUEGO DE ADIVINANZA 🎲")

# Generar un número aleatorio entre 1 y 100
numero_secreto = random.randint(1, 100)

adivinado = False

while not adivinado:
    intento = int(input("Adivina el número (entre 1 y 100): "))

    if intento < numero_secreto:
        print("El número secreto es MÁS alto.")
    elif intento > numero_secreto:
        print("El número secreto es MÁS bajo.")
    else:
        print("🎉 ¡Correcto! Has adivinado el número.")
        adivinado = True
```
